## <font color='green'>INPUT DATA</font>
The one provided here has been renamed to 'flipkart_data.csv'. The provided dataset must be stored in the dataset sub-directory. 

```python
name_of_dataset = "flipkart_data.csv"
```

## <font color='green'>FEATURE SELECTION</font>
To select a subset of attributes from the entire dataset, while keeping all the records. This narrows down the observation spectrum and helps avoid inferring ambiguous relations with other unimportant attributes. The fields returned by this model are ['pid','name','product category tree','brand','description']. The procedure has been explained and implemented in <b>feature_selection.ipynb</b>

```python
data = perform_feature_selection(name_of_dataset)

#feature selection accesses the dataset in its method definition ------
#returns only the relevant attributes from the dataset ----------------------
```

## <font color='green'>DISCARD NaN ENTRIES</font>
Dropping the records where the description field or the category field is returned to be empty or a non-comprehensible value.

```python
data = data.dropna(subset=["description","cat"])

# drop any rows with a NaN value in the description or 
# category columns of the dataset -------------------------------------------------------

data = data.reset_index(drop=True) #reset indices of the dataset --------------
```

## <font color='green'>OBTAIN PRIMARY CATEGORIES</font>
The primary category tree was splitted on the string ' >> ' in order to obtain the list of category tree hierarchy. The procedure has been explained and implemented in <b>retrieve_categories.ipynb</b>

```python
data,categories = retrieve_primary_categories(data) 

#identify and shortlist appropriate level-0 (primary) categories
#returns the updated dataset along with the categories in a separate list for further reference----------
```

### ordinal encoding the shortlisted primary categories
Can be utilized by the SVM model approach later down in the implementation pipeline.

```python
OE_categories = {k: v for v, k in enumerate(categories)}
with open("output_files/categories.json", "w") as outfile: 
    json.dump(OE_categories, outfile)

#creating ordinal encoded JSON file for the shrotlisted primary categories
```

## <font color='green'>PREPROCESSING</font>
Preprocessing plays a pivotal role in filtering out uneccessary text segments and refining the overall performance of the classifier. Multiple steps together constitute the preprocessing pipeline, which has been explained and implemented in <b>description_preprocessing.ipynb</b> 

```python
data = preprocess_description(data) # preprocess the dataset
```

## visualizing preprocessed category wise record counts in the dataset

```python
vis_data(data) #for visualization of the category-wise distribution of records
```

## computing feature represenations to be utilized by the SVM model
Pre-trained feature representations have been utilized in order to design the feature set for the SVM model later in the pipeline. The procedure has been explained and implemented in <b>compute_embeddings.ipynb</b>

```python
glove_description = embedding_model(data['custom'])

#computing embeddings for the preprocessed descriptions
# returns a dataframe for further reference ---------------------------------
```

## <font color='green'>CALL TO BAG-OF-WORDS MODEL</font>
To create the bag of words and predict the categories for the records in the test set. The procedure has been explained and implemented in <b>bag_of_words.ipynb</b>

```python
bow_model(data,categories) 

#call to the bag-of-words model for creating the BOWs and computing results --
# the results are stored as a CSV file
```

## <font color='green'>CALL TO SUPPORT VECTOR MACHINES (SVM) MODEL</font>
To use the SVC class of SVM supervised learning methods in order to classify the test set descriptions. the procedure has been further explained and implemented in <b>svm_model.ipynb</b>

```python
svm_results = svm_main(glove_description['embedding'],data['cat'],categories) #call to the SVM
```

## storing SVM results 

```python
key_list = list(OE_categories.keys())
val_list = list(OE_categories.values())
for i in range(len(svm_results)):
    position = svm_results.loc[i,'actual']
    svm_results.loc[i,'actual'] = key_list[position]
    position = svm_results.loc[i,'predicted']
    svm_results.loc[i,'predicted'] = key_list[position]
svm_results.to_csv("output_files/string_svm_results.csv",index=False)
```
